```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G08 = nx.read_edgelist(
    './data/p2p-Gnutella08.txt',
    comments='#',
    delimiter=None,
    create_using=nx.DiGraph,
    nodetype=None,
    data=True,
    edgetype=None,
    encoding='utf-8',
)

# Layout: increase repulsion a bit and scale coordinates up
k08 = 2.5 / (G08.number_of_nodes() ** 0.5)
pos08 = nx.spring_layout(G08, seed=42, iterations=250, k=k08, scale=8.0)
```

```python
plt.figure(figsize=(14, 10), dpi=160)
ax = plt.gca()
ax.set_axis_off()

# Nodes: tiny + very transparent (reduces overplotting)
nx.draw_networkx_nodes(
    G08,
    pos08,
    node_size=6,
    node_color="#6baed6",
    alpha=0.10,
    linewidths=0,
)

# Edges: hairline + extremely transparent; arrowheads kept but tiny
nx.draw_networkx_edges(
    G08,
    pos08,
    arrows=True,
    arrowstyle="-|>",
    arrowsize=4,
    width=0.25,
    alpha=0.03,
    edge_color="black",
    connectionstyle="arc3,rad=0.05",
)

plt.tight_layout()
```
